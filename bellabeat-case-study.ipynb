##### 3.2.2.1 Hourly Intensities
* Initially I counted activity minute for each activity level i.e. sedentary active activity, lightly active activity, fairly active activity, and very active activity by using the following SQL queries:

`SELECT` Id, ActivityDate, Hour, `COUNT`(intensity) `AS` SedentaryMinutes `FROM` bigqueryintroduction.bellabeat.intensities `WHERE` intensity =0 `GROUP BY` Id, ActivityDate, Hour `ORDER BY` Id, ActivityDate, Hour

* To get Total Intensity= (SedentaryMinutes*0 + LightlyActiveMinutes*1 + FairlyActiveMinutes*2 + VeryActiveMinutes*3) and Average Intensity = Total Intesity/60 , I used the following query:

`SELECT` Id, ActivityDate, Time, `SUM`(Intensity) `AS` TotalIntensity, avg(Intensity) `AS` AverageIntensity `FROM` bigqueryintroduction.bellabeat.intensities `GROUP BY` Id, ActivityDate, Time, Hour `ORDER BY` Id, ActivityDate, Time

* For hourly intensities dataset, I joined the results produced by above queries using spreadsheet.

#### 3.2.5 METs
Active minutes are calculated through **metabolic equivalents** (METs). This metric, which factors in the user's body mass, is used to estimate exercise intensity. METs are calculated through a ratio (source):

**Rate of energy expended during an activity : rate of energy expended during rest**

1 MET is the rate of energy a user expends during rest or sitting quietly, while a user typically expend 3 METs or more during a moderately-intense activity such as walking.superscript^1^

Based on **healthline.com**[link](https://www.healthline.com/health/what-are-mets#examples), the MET values are assigned as **Light < 3.0 METs, Moderate 3.0–6.0 METs, Vigorous > 6.0 METs.**superscript^2^

**I have considered LightMetMinutes for users who have METs< 3.0, ModerateMetMinutes for users who have METs between 3.0–6.0, VigorousMetMinutes for users who have METs>6.0.**

I have uploaded the Minute METs (narrow) data file on BigQuery with table name "met" in project "bellabeat" and then used SQL queries to find daily and hourly activity minutes based on METs.

##### 3.2.5.1 Hourly METs
To determine activity minutes based on METs, shown below are the sql queries :

`SELECT` Id, ActivityDate, Time, `COUNT`(METs) `AS` LightMetMinutes FROM bigqueryintroduction.bellabeat.met `WHERE` ActualMET<3 `GROUP BY` Id, ActivityDate, Time `ORDER BY` Id, ActivityDate, Time

`SELECT` Id, ActivityDate, Time, `COUNT`(METs) `AS` ModerateMetMinutes, `SUM`(ActualMET) `AS` ModerateEquivalentMETs `FROM` bigqueryintroduction.bellabeat.met `WHERE` ActualMET>=3 `AND` ActualMET<=6 `GROUP BY` Id, ActivityDate, Time `ORDER BY` Id, ActivityDate, Time

`SELECT` Id, ActivityDate, Time, `COUNT`(METs) `AS` VigorousMetMinutes, `SUM`(ActualMET) `AS` VigorousEquivalentMETs `FROM` bigqueryintroduction.bellabeat.met `WHERE` ActualMET>6 `GROUP BY` Id, ActivityDate, Time `ORDER BY` Id, ActivityDate, Time

##### 3.2.5.2 Daily METs
Similar to the Hourly METs queries, I used the following queries to determine daily active activity minutes :

`SELECT` Id, ActivityDate, `COUNT`(METs) `AS` LightMetMinutes FROM bigqueryintroduction.bellabeat.met `WHERE` ActualMET<3 `GROUP BY` Id, ActivityDate `ORDER BY` Id, ActivityDate

`SELECT` Id, ActivityDate, `COUNT`(METs) `AS` ModerateMetMinutes, `SUM`(ActualMET) `AS` ModerateEquivalentMETs `FROM` bigqueryintroduction.bellabeat.met `WHERE` ActualMET>=3 `AND` ActualMET<=6 `GROUP BY` Id, ActivityDate `ORDER BY` Id, ActivityDate

`SELECT` Id, ActivityDate, `COUNT`(METs) `AS` VigorousMetMinutes, `SUM`(ActualMET) `AS` VigorousEquivalentMETs `FROM` bigqueryintroduction.bellabeat.met `WHERE` ActualMET>6 `GROUP BY` Id, ActivityDate `ORDER BY` Id, ActivityDate

> **Using spreadsheets, I used the "sum" formula to determine EquivalentMETs, EquivalentMETs = VigorousEquivalentMETs + ModerateEquivalentMETs.**

### 4.1 Hourly Calories
By using hourly calories dataset, I compared calories burned on certain day with respective time periods.

```r
## hourly calories
hr_calories<-hr_calories %>%
  mutate(WeekDay = weekdays(ActivityDate))
hr_calories$Time<-as.character(hr_calories$Time)
ggplot(hr_calories,aes(x=Time,y=Calories))+geom_histogram(stat="identity", fill= "tan3")+theme(axis.text.x = element_text(angle = 90))+labs(title="Calories Burned Vs Time")

### calculating average calories burned
c1<-hr_calories%>%
  group_by(WeekDay)%>%
  summarise(AverageCalories = mean(Calories))
ggplot(c1, aes(x=WeekDay,y=AverageCalories))+geom_bar(stat="identity", fill = "orange2")+geom_text(aes(label = round(AverageCalories)),nudge_y = 2)+labs(title="Average Calories Burned Vs WeekDay")
```

### 4.6 Merging Datasets

I have merged(outer join) daily activity, daily METs, and Sleep Info datasets into a new data frame named as merge. Then I created a new column for the Week Day.

Two new variables were also created and were calculated based on :

* **ActiveMinutes -** Combined active activity minutes i.e. (Lightly + Fairly + Very) Active Minutes.

* **METMinutes -** Combination of moderate intense physical activity minutes and vigorous intense activity minutes. 

```r
## merging daily activity datset, daily METs dataset, daily sleep dataset :
merge1<-merge(activity,MET, all = TRUE)
merge2<-merge(merge1,sleep,all = TRUE)

### calculating day of the week, Active Minutes, and MET Minutes:
merge2$WeekDay<-weekdays(merge2$ActivityDate)
merge2$ActiveMinutes<-merge2$LightlyActiveMinutes+merge2$FairlyActiveMinutes+merge2$VeryActiveMinutes
merge2$METMinutes<-merge2$ModerateMetMinutes+merge2$VigorousMetMinutes
```

I created a new variable which categorise every dataset recorded into cateogories of type of user activity. These physical activities categories were : Inactive,Not Enough, Active, and Very Active.superscript^4^

* **Inactive** - is not getting any moderate level or vigorous intensity physical activity beyond basic movement from daily life activities.

* **Not Enough** - is doing some moderate- or vigorous-intensity physical activity but less than 150 minutes of moderate-intensity physical activity a week or 75 minutes of vigorous-intensity physical activity or the equivalent combination. This level is less than the target range for meeting the key guidelines for adults.

* **Active** - is doing the equivalent of 150 minutes to 300 minutes of moderate-intensity physical activity a week. This level meets the key guideline target range for adults.

* **Very Active** - is doing the equivalent of more than 300 minutes of moderate-intensity physical activity a week. This level exceeds the key guideline target range for adults.

> **Required moderate-intensity activity minutes are 150-300 each week and atleast 75 minutes minimum of vigorous physical activity each week, I have aggregated this information to 180-360 active minutes on daily basis, and required 500 MET minutes of activity each week to atleast 70 MET minutes of daily activity.**

> **I have also categorised sleep into :-  Good Sleep between 6 hrs to 9 hrs, Less Sleep for less than 6 hrs of sleep, and Over Sleep for more than 9 hrs of sleep.**

```r
### categorising the activeness based on Activity minutes and MET minutes, and categoorising sleep patterns based on time asleep :
merge<-merge2%>%
  mutate(UserActiveness = factor(case_when(ActiveMinutes==0|METMinutes==0 ~ "Inactive", ActiveMinutes<180|METMinutes<70~"Not Enough",(ActiveMinutes>=180 & ActiveMinutes<360)|(METMinutes>=70 & METMinutes<140) ~"Active",ActiveMinutes>=360|METMinutes>=140 ~"Very Active")))
merge<-merge%>%
  mutate(SleepInfo=factor(case_when(TimeAsleep>=540~"Over Sleep",TimeAsleep<540 & TimeAsleep>=360~"Good Sleep",TimeAsleep<360~"Less Sleep")))
```
